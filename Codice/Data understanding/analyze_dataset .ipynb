```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

TRAIN_PATH = r""
TEST_PATH = r""

x_train = pd.read_csv(TRAIN_PATH, header=None)
x_test = pd.read_csv(TEST_PATH, header=None)
y_train = x_train.iloc[:, 0]
x_train = x_train.drop(0, axis=1)
y_test = x_test.iloc[:, 0]
x_test = x_test.drop(0, axis=1)

x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5, random_state=42)
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x=y_train).set_title("train")
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x=y_val).set_title("val")
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x=y_test).set_title("test")
```
