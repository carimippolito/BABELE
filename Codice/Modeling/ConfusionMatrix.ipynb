```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

CSV_MANHATTAN_INPUT_PATH = r""

x_test_m28 = pd.read_csv(CSV_MANHATTAN_INPUT_PATH +
                         r"\28 distanze\test.csv", header=None)
y_test_m28 = x_test_m28.iloc[:, 0]
x_test_m28 = x_test_m28.drop(0, axis=1)

y_test_m28 = y_test_m28.to_numpy(np.int32)
x_test_m28 = x_test_m28.to_numpy(np.float32)
```

```python
from joblib import load
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

MODEL_INPUT_PATH = r""

labels = np.array(["Italian", "English", "German", "Spanish", "Dutch", "Russian"])

pipe = load(MODEL_INPUT_PATH + "\\ml\\v1.0\\filename.joblib")

prediction = pipe.predict(x_test_m28)

C = confusion_matrix(labels[prediction - 1], labels[y_test_m28 - 1])
disp = ConfusionMatrixDisplay(confusion_matrix=C, display_labels=labels)

fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Confusion Matrix\n", fontsize=20)
disp.plot(ax=ax, cmap=plt.cm.Blues, xticks_rotation=45)
```
